```python
import os 
import argparse
from utils.functions import get_output_dir_from_args
from utils.parser import add_shared_parser_arguments

import matplotlib.pyplot as plt
```

```python
MODEL = "cnn"
DATASET = "mnist"
NUM_EPOCHS = "40"

parser = argparse.ArgumentParser(description="Visualize data")
add_shared_parser_arguments(parser)
args = parser.parse_args(args=["--model", MODEL, "--dataset", DATASET, "--num_epochs", NUM_EPOCHS])

output_dir = get_output_dir_from_args(args, create=False)
```

```python
os.listdir("../" + output_dir)
```

```python
import json

with open("../" + output_dir + "/training_stats.json", "r") as f:
    stats = json.load(f)
```

```python
training_loss = [data["training_loss"] for data in stats]
validation_loss = [data["validation_loss"] for data in stats]
epochs = list(range(1, 41))
```

```python

plt.title("Training and Validation Loss")
plt.plot(epochs, training_loss, label="train")
plt.plot(epochs, validation_loss, label="val")
plt.xlabel("Epochs")
plt.ylabel("Average loss/sample")


plt.show()
```

```python
accuracy = [data["validation_accuracy"] for data in stats]

plt.plot(epochs, accuracy)
plt.title("Accuracy")
plt.xlabel("Epochs")
```
